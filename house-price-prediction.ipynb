### `4. Data Preprocessing`

```python
X = df.drop(columns=[TARGET_COL])

y = df[TARGET_COL]
```

```python
X.head()
```

```python
y.head()
```

```python
# Train test Split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    random_state=RANDOM_STATE
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Preprocessing Pipeline

```python
numerical_features = X_train.select_dtypes(include=(np.number)).columns.tolist()
print( "Numerical Features", numerical_features)
```

```python
categorical_features = X_train.select_dtypes(exclude=(np.number)).columns.tolist()
print("Categorical Features", categorical_features)
```

```python
# Numerical Features - preprocessing steps
numerical_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy='median')), ("scaler", StandardScaler())]
)

# Categorical Features - preprocessing steps
categoical_transformer = Pipeline(
    steps=[("imputer", SimpleImputer(strategy='most_frequent')), ("onehot", OneHotEncoder(handle_unknown="ignore")) ]
)
```

```python
# Preprocessing Pipeline

preprocess = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categoical_transformer, categorical_features)
    ]
)


```

## 5. `Base Line Model ( No CV, No Tuning)`

```python
baseline_pipeline = Pipeline(
    steps=[
        ('preprocess', preprocess),
        ('model', LinearRegression())
    ]
)
```

```python
# preprocess the data 

baseline_pipeline.fit(X_train, y_train)
```

### Evaluation of baseline model

```python
train_baseline_prediction = baseline_pipeline.predict(X_train)
test_baseline_prediction = baseline_pipeline.predict(X_test)
```

```python
train_baseline_rmse = root_mean_squared_error(y_train, train_baseline_prediction)
train_baseline_mae = mean_absolute_error(y_train, train_baseline_prediction)
train_baseline_r2 = r2_score(y_train, train_baseline_prediction)
# train_baseline_acc = accuracy_score(y_train, train_baseline_prediction)
```

```python
print("\n TRAIN BASELINE MATRICS (LinearRegression)")
print("RMSE", train_baseline_rmse)
print("MAE", train_baseline_mae)
print("R2", train_baseline_r2)
# print("Accuracy score", train_baseline_acc)
```

```python
test_baseline_rmse = root_mean_squared_error(y_test, test_baseline_prediction)
test_baseline_mae = mean_absolute_error(y_test, test_baseline_prediction)
test_baseline_r2 = r2_score(y_test, test_baseline_prediction)
# test_baseline_acc = accuracy_score(y_test, test_baseline_prediction)
```

```python
print("============== TEST BASELINE MATRICS (LinearRegression) ==============================")
print("RMSE", test_baseline_rmse)
print("MAE", test_baseline_mae)
print("R2", test_baseline_r2)
# print("Accuracy score", test_baseline_acc)
```

## `6. Model Selection and Optimization`

```python
# models to try

models = {
    "LinearRegression": LinearRegression(),
    "Ridge":Ridge(random_state=RANDOM_STATE),
    "Lasso":Lasso(random_state=RANDOM_STATE, max_iter=10000),
    "RandomForest": RandomForestRegressor(),
    "HistGB": HistGradientBoostingRegressor(),
}
```

```python
k = 5
# cross validatoin
cv = KFold(n_splits=k, shuffle=True, random_state=RANDOM_STATE)
```

```python
scoring = {
    'rmse': "neg_root_mean_squared_error",
    "mae":"neg_mean_absolute_error",
    "r2": "r2"
}
```

```python
rows = []

for name, model in models.items():
    # print(name, model)
    pipe = Pipeline(
        steps=[
            ('preprocess', preprocess),
            ('model', model)
        ]
    )

    scores = cross_validate(pipe, X_train, y_train, cv=cv, scoring=scoring, n_jobs=1)
    rows.append({
        "model": name,
        "cv_rmse": -scores['test_rmse'].mean(),
        "cv_mae": -scores['test_mae'].mean(),
        "cv_r2": -scores['test_r2'].mean(),
    })


# sort based on lowest rmse value

cv_results = pd.DataFrame(rows).sort_values('cv_rmse')
print("============== CV Model Comparison =======================")
print(cv_results)
```
